## Project Overview:
This notebook performs a comprehensive analysis of road accident data to identify key trends and patterns.
The analysis includes descriptive statistics and data visualizations, addressing various aspects such as:
    
- Frequency of accidents over time (yearly, monthly, daily, and hourly)
- Geographical distribution and accident hotspots
- Accident severity distribution
- Environmental and road conditions
- Vehicle and driver-related insights
- Temporal patterns (weekdays vs. weekends, peak hours)
- Contributing factors to accidents
- Injury and fatality analysis
- Comparative analysis between urban and rural areas

The dataset used contains information about accidents, including the accident date, time, location, severity, weather, road type, and vehicle type involved in the accidents

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
df = pd.read_csv("C:\\Users\\Admin\\Downloads\\Road Accident Data.csv")

# Display basic info
display(df.info())
display(df.head())
```

```python
# Handle missing values
df.dropna(inplace=True) 

# Convert date columns to datetime format
df['Accident Date'] = pd.to_datetime(df['Accident Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.hour
```

### Data Loading and Cleaning:
The dataset has been loaded successfully, and basic information about the dataset has been displayed.Rows with missing values 
have been removed using `df.dropna()`. Additionally, the `Accident Date` column has been converted into a datetime format, 
and the `Time` column has been extracted into hours for better temporal analysis.


```python
# 1. Frequency of Accidents Over Time
print(f"Total number of accidents recorded: {df.shape[0]}")

accidents_by_year = df.groupby('Year').size()
accidents_by_month = df.groupby('Month').size()
accidents_by_day = df.groupby('Day_of_Week').size()
accidents_by_hour = df.groupby('Time').size()

plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
accidents_by_year.plot(kind='bar', title='Accidents per Year')

plt.subplot(2, 2, 2)
accidents_by_month.plot(kind='bar', title='Accidents per Month')

plt.subplot(2, 2, 3)
accidents_by_day.plot(kind='bar', title='Accidents by Day of the Week')

plt.subplot(2, 2, 4)
accidents_by_hour.plot(kind='bar', title='Accidents by Hour')

plt.tight_layout()
plt.show()
```

### 1. Frequency of Accidents Over Time:

- **Total Number of Accidents**: The dataset contains **300492** of accidents.
- **Accidents per Year**: The number of accidents shows a **gradual decrease** over the year.
- **Accidents per Month**: Accidents are more frequent in **November**, possibly due to adverse weather conditions.
- **Accidents by Day of the Week**: The lowest number of accidents occur on **Sundays**.
- **Accidents by Hour**: Most accidents occur during **rush hours**, or office hours i.e **8AM** and **5PM**.


```python
# 2. Geographical Distribution
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Longitude'], y=df['Latitude'], alpha=0.3)
plt.title("Accident Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Identify accident hotspots
top_locations = df['Local_Authority_(District)'].value_counts().head(10)
plt.figure(figsize=(10, 5))
top_locations.plot(kind='bar', color='red')
plt.title("Top 10 Accident-Prone Locations")
plt.xlabel("Location")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()
```

### 2. Geographical Distribution:
- **Accident Locations**: The scatter plot shows the distribution of accidents across different geographical areas. 
- **Top 10 Accident-Prone Locations**: The **top 10 locations** with the highest frequency of accidents are shown in the bar chart. These locations include **Birmingham**, and **Leeds**.


```python
# 3. Accident Severity Analysis
severity_counts = df['Accident_Severity'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(8, 5))
ax = severity_counts.plot(kind='bar', color=['red', 'orange', 'green'])

# Adding percentage labels on top of each bar
total = severity_counts.sum()
for p in ax.patches:
    height = p.get_height()
    percentage = (height / total) * 100
    ax.annotate(f'{percentage:.2f}%', 
                (p.get_x() + p.get_width() / 2, height), 
                ha='center', va='bottom', fontsize=12)

plt.title("Accident Severity Distribution")
plt.xlabel("Severity")
plt.ylabel("Count")
plt.show()
```

### 3. Accident Severity Analysis:
- **Accident Severity Distribution**: The majority of accidents are **Minor**, with a smaller portion being **Serious** or **Fatal**.
- **Fatal/Serious Accidents**: **14.64%** of accidents resulted in **fatal/serious** injuries.


### 4. Demographic Insights

Dataset doesn't has Age & Gender columns

```python
# 5. Environmental and Road Conditions
weather_counts = df['Weather_Conditions'].value_counts()
plt.figure(figsize=(8, 5))
weather_counts.plot(kind='bar')
plt.title("Accidents by Weather Condition")
plt.show()

road_counts = df['Road_Type'].value_counts()
plt.figure(figsize=(8, 5))
road_counts.plot(kind='bar', color='purple')
plt.title("Accidents by Road Type")
plt.show()
```

### 5. Environmental and Road Conditions:
- **Weather Conditions**: Accidents are more frequent during **dry** weather conditions.
- **Road Types**: The majority of accidents occur on **single carriageway**.


```python
# 6. Vehicle and Driver Information
vehicle_counts = df['Vehicle_Type'].value_counts()
plt.figure(figsize=(8, 5))
vehicle_counts.plot(kind='bar', color='skyblue')
plt.title("Accidents by Vehicle Type")
plt.show()
```

### 6. Vehicle and Driver Information:
- **Vehicle Types**: The most common vehicles involved in accidents are **cars**.
- **Vehicle Type and Severity**: Larger vehicles like **trucks** tend to be involved in more severe accidents.


```python
# 7. Temporal Patterns
weekend_accidents = df[df['Day_of_Week'].isin(['Saturday', 'Sunday'])].shape[0]
weekday_accidents = df[~df['Day_of_Week'].isin(['Saturday', 'Sunday'])].shape[0]
plt.pie([weekday_accidents, weekend_accidents], labels=['Weekdays', 'Weekends'], autopct='%1.1f%%', colors=['blue', 'orange'])
plt.title("Accident Distribution: Weekdays vs. Weekends")
plt.show()
```

### 7. Temporal Patterns:
- **Weekdays vs. Weekends**: There is a higher number of accidents on **weekdays** compared to weekends.


```python
# 8. Contributing Factors
factor_counts = df['Junction_Control'].value_counts().head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 5))
ax = factor_counts.plot(kind='bar', color='purple')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)  # Adjust the rotation angle and font size
plt.title("Top Contributing Factors to Accidents")
plt.ylabel("Number of Accidents")

# Adjust the layout to make space for the labels
plt.tight_layout()

plt.show()
```

### 8. Contributing Factors:
- **Top Contributing Factors**: The most common contributing factor is **Uncontrolled junctions or give way**.


```python
# 9. Injury and Fatality Analysis
injury_counts = df['Number_of_Casualties'].value_counts().head(15)  # Limit to top 15 categories

# Plotting the bar chart
plt.figure(figsize=(10, 6))
ax = injury_counts.plot(kind='bar', color='red')

# Adding percentage labels on top of each bar
total = injury_counts.sum()  # Total number of casualties
for p in ax.patches:
    height = p.get_height()
    percentage = (height / total) * 100
    ax.annotate(f'{percentage:.2f}%',  # Format the percentage to two decimal places
                (p.get_x() + p.get_width() / 2, height),  # Position of the text
                ha='center', va='bottom', fontsize=12)  # Aligning and formatting the text

plt.title("Injury and Fatality Distribution")
plt.xlabel("Number of Casualties")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

### 9. Injury and Fatality Analysis:
- **Injury/Fatality Distribution**: The highest number of injuries are **minor**, with a smaller number of **fatalities**.


```python
# 10. Comparative Analysis
urban_accidents = df[df['Urban_or_Rural_Area'] == 'Urban'].shape[0]
rural_accidents = df[df['Urban_or_Rural_Area'] == 'Rural'].shape[0]
plt.pie([urban_accidents, rural_accidents], labels=['Urban', 'Rural'], autopct='%1.1f%%', colors=['green', 'brown'])
plt.title("Accident Distribution: Urban vs. Rural")
plt.show()
```

### 10. Comparative Analysis:
- **Urban vs Rural Areas**: Accidents are more frequent in **urban areas** than in **rural areas**.
